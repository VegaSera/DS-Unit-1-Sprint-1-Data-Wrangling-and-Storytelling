##Modify and Create Columns using .apply()

```python
#Look at top 20 employment titles.
df['emp_title'].value_counts(dropna=False)[:20]
#This reveals there are several redunant jobs, several missing jobs. Column needs to be cleaned.
```

```python
#Number of unique titles
len(df['emp_title'].value_counts(dropna=False))
```

```python
#Define how we want to clean up the data, and what data we can expect.
#We know we can expect uncapitalized titles (like manager and Manager) 
#Leading or trailing spaces can be included and will show as a different title.
#Null values are present and will throw an error if we try to use a string method on a float.
examples = ['owner', 'Supervisor ', ' Project Manager', np.NaN]
```

```python
#Defining a function to clean up the example, and our column.
def clean_title(job):
  if isinstance(job, str): #Makes sure the job is not NaN, and if it is, treats it differently.
    return job.title().strip()
  else:
    return "Unknown" #Returns a string instead of NaN again, so the entire column is now string objects.
```

```python
#List comprehension like we've done above, to clean the examples.
[clean_title(example) for example in examples]
```

```python
#Time to apply our functions to our column of job titles
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```python
#checking to see how much of a difference our cleaning made
df['emp_title'].value_counts(dropna=False)[:20]
```

```python
len(df['emp_title'].value_counts())
```

```python
#Testing to see if any null values remain.
df['emp_title'].isnull().sum()
```

```python
#We know our data can only either be 36 months or 60 months.
def months_to_int(months):
  return int(months.strip()[:2]) #Takes the statement, strips the whitespace, and only returns the first two characters, which will either be 36 or 60.
```

```python
def loan_standing(status):
  if status == "Current" or status == "Fully Paid":
    return 1
  else:
    return 0
```
